```python
import requests
import time
import random
import json

# ThingsBoard instance URL
THINGSBOARD_HOST = "http://demo.thingsboard.io"

# Access tokens for your virtual devices
ACCESS_TOKENS = ["4GmWfJsKZn7MUfrgMml9", "EEMmgWkKqYjL3K4HBC8c", "npyn8RFMNUtLTxGPE2dg"]

# Virtual Devices
VIRTUAL_DEVICES = [
     {"name": "TCV D1", "token": ACCESS_TOKENS[0]},
     {"name": "TCV D2", "token": ACCESS_TOKENS[1]},
     {"name": "TCV D3", "token": ACCESS_TOKENS[2]},

]

# HTTP POST headers
HEADERS = {"Content-Type": "application/json"}

# Data simulation function
def simulate_data(device):
    """
    Generate simulated data for a device.
    """
    return {
            "max_temperature": round(random.uniform(50, 80), 2),
            "min_temperature": round(random.uniform(10, 30), 2),
            "max_current": round(random.uniform(30, 50), 2),
            "min_current": round(random.uniform(0.5, 10), 2),
            "min_voltage": round(random.uniform(220, 240), 2),
            "max_voltage": round(random.uniform(240, 400), 2),
    }

# Data publishing function
def publish_data():
    """
    Publish simulated data for all virtual devices.
    """
    for device in VIRTUAL_DEVICES:
        telemetry_data = simulate_data(device)
        print(json.dumps(telemetry_data))

        # Send telemetry data
        url = f"{THINGSBOARD_HOST}/api/v1/{device['token']}/telemetry"
        response = requests.post(url, headers=HEADERS, data=json.dumps(telemetry_data))

        if response.status_code == 200:
            print(f"Telemetry sent for {device['name']}: {telemetry_data}")
        else:
            print(f"Failed to send telemetry for {device['name']} - Status: {response.status_code}")

# Main loop
if __name__ == "__main__":
    try:
        print("Starting virtual device simulation...")
        while True:
            publish_data()
            time.sleep(5)  # Send data every 5 seconds
    except KeyboardInterrupt:
        print("Simulation stopped.")
```
